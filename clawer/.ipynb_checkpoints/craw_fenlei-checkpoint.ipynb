```python
from bs4 import BeautifulSoup
from urllib.error import HTTPError, URLError
import urllib.request
import time
import socket
from collections import deque
import pickle
```

```python
socket.setdefaulttimeout(30)
all_fenlei = set()
work_list = deque()
```

```python
def get_html(url, log_file, code='utf-8'):
    """获取请求url返回的页面，默认utf-8解码"""
    for i in range(3):
        try:
            page = urllib.request.urlopen(url)
            break
        except HTTPError as e:
            print('!!!%s，服务器不能应答，Error Code:%s' % (url, str(e.code)))
            print('!!!%s，服务器不能应答，Error Code:%s' %
                  (url, str(e.code)), file=log_file)
        except URLError as e:
            print('!!!%s，连接服务器失败，Reason:%s' % (url, str(e.reason)))
            print('!!!%s，连接服务器失败，Reason:%s' %
                  (url, str(e.reason)), file=log_file)
        except socket.timeout:
            print('!!!%s 访问超时' % url)
            print('!!!%s 访问超时' % url, file=log_file)
            time.sleep(60)
        except Exception as e:
            print('!!!%s 访问出错' + str(e) % url)
            print('!!!%s 访问出错' + str(e) % url, file=log_file)
        time.sleep(20)
    else:
        print('!!!%s 页面访问失败，丢弃' % url)
        print('!!!%s 页面访问失败，丢弃' % url, file=log_file)
        return ""

    try:
        html = page.read().decode(code, errors='ignore')
        return html
    except:
        print('!!!%s 页面读取失败，丢弃' % url)
        print('!!!%s 页面读取失败，丢弃' % url, file=log_file)
        return ""
```

```python
def get_in_fenlei(fenlei, fenlei_file, log_file):
    fenlei_quote = urllib.parse.quote(fenlei)
    fenlei_url = 'http://baike.baidu.com/fenlei/%s' % fenlei_quote
    html = get_html(fenlei_url, log_file)
    if not html:
        print('FAILED:%s' % fenlei)
        print('FAILED:%s' % fenlei, file=log_file)
        return None

    # 获取下级分类和相关分类
    soup = BeautifulSoup(html, 'html.parser')
    children_and_brother = soup.find_all(
        'a', {'class': 'child-link nslog:7449'})
    for f in children_and_brother:
        text = f.get_text()
        if text not in all_fenlei:
            all_fenlei.add(text)
            work_list.append(text)

    # 获取一个分类中列举词条上所依附的开放分类
    index = 1
    while True:
        url = 'http://baike.baidu.com/fenlei/%s?limit=30&index=%d&offset=%d#gotoList' % (
            fenlei_quote, index, 30 * (index - 1))
        html = get_html(url, log_file)
        if not html:
            print("???Can't get this page:%s" % url)
            print("???Can't get this page:%s" % url, file=log_file)
            index += 1
            continue

        soup = BeautifulSoup(html, 'html.parser')
        link_category = soup.find_all(
            'a', {'class': 'link-category nslog:7451'})
        for f in link_category:
            text = f.get_text()
            if text not in all_fenlei:
                all_fenlei.add(text)
                work_list.append(text)

        # 如果没有下一页了，就跳出
        if not soup.find('a', {'class': 'next nslog:7453'}):
            break

        index += 1

    fenlei_file.write(fenlei + '\n')
    print('SUCCESS:%s, all_fenlei size:%d, rest in work_list:%d' %
          (fenlei, len(all_fenlei), len(work_list)))
    print('SUCCESS:%s, all_fenlei size:%d, rest in work_list:%d' %
          (fenlei, len(all_fenlei), len(work_list)), file=log_file)
```

```python
# 第一次启动，从种子开始
seeds = [
    '政治人物',
    '历史人物',
    '文化人物',
    '虚拟人物',
    '经济人物',
    '动物',
    '植物',
    '自然灾害',
    '自然资源',
    '自然现象',
    '美术',
    '戏剧',
    '舞蹈',
    '摄影',
    '曲艺',
    '书画',
    '建筑',
    '语言',
    '体育组织',
    '体育奖项',
    '体育设施',
    '体育项目',
    '组织机构',
    '政治',
    '军事',
    '法律',
    '民族',
    '交通',
    '经济',
    '党务知识',
    '各国历史',
    '历史事件',
    '历史著作',
    '文物考古',
    '行政区划',
    '地形地貌',
    '科研机构',
    '互联网',
    '航空航天',
    '医学',
    '电子产品',
    '动漫',
    '电影',
    '电视剧',
    '小说',
    '电视节目',
    '演出',
    '美容',
    '时尚',
    '旅游'
]

for seed in seeds:
        all_fenlei.add(seed)
        work_list.append(seed)
```

```python
# 启动程序，如果是第一次，要先执行上面的种子，如果不是第一次，要先执行下面的，得到对象
cnt = 0
with open('fenlei.log', 'at', encoding='utf-8') as log_file:
    with open('fenlei.txt', 'at', encoding='utf-8') as fenlei_file:
        # 每次启动在log中写上日期
        from datetime import datetime
        print('-'*25,datetime.today(),'-'*25, file=log_file)
        
        while work_list:
            get_in_fenlei(work_list.popleft(), fenlei_file, log_file)
            # 每进行10个分类，就把对象序列化转储文件中
            cnt += 1
            if cnt % 10 == 0:
                with open('all_fenlei', 'wb') as f:
                    pickle.dump(all_fenlei, f)
                with open('work_list', 'wb') as f:
                    pickle.dump(work_list, f)
```

```python
# 之后启动，从保存的对象开始，最后不管程序运行到哪一步，都可以从序列化对象中得到相关数据
with open('all_fenlei', 'rb') as all_fenlei_file:
    all_fenlei = pickle.load(all_fenlei_file)
with open('work_list', 'rb') as work_list_file:
    work_list = pickle.load(work_list_file)
```

```python
# 爬取过程all_fenlei和work_list的size变化数据（从log中获取）：
import re
all_fenlei_cnt = []
work_list_cnt = []

with open('fenlei.log', 'r', encoding='utf-8') as log:
    lines = log.readlines()
    pattern = re.compile('.*size:([0-9]+).*list:([0-9]+)')
    
    for line in lines:
        if line[0] == 'S':
            group = pattern.match(line).groups()
            all_fenlei_cnt.append(int(group[0]))
            work_list_cnt.append(int(group[1]))
    
    print('all_fenlei_cnt:',len(all_fenlei_cnt),'work_list_cnt:',len(work_list_cnt))
```

```python
# 画出all_fenlei和work_list的size的变化图
import matplotlib.pyplot as plt
%matplotlib inline

x = range(len(all_fenlei_cnt))
plt.plot_date(x, all_fenlei_cnt)
plt.plot(x, work_list_cnt)
```
